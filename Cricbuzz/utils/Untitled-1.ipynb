```python
import requests
import pandas as pd
from sqlalchemy import create_engine
from urllib.parse import quote_plus

# ----------------------------
# Database Connection
# ----------------------------

DB_USER = "root"
DB_PASSWORD = quote_plus("Vasu@76652")
DB_HOST = "localhost"
DB_PORT = 3306
DB_NAME = "cricketdb"

engine = create_engine(
    f"mysql+pymysql://{DB_USER}:{DB_PASSWORD}@{DB_HOST}:{DB_PORT}/{DB_NAME}"
)

print("Database connected")

# ----------------------------
# Get Match IDs from matches table
# ----------------------------

query = "SELECT Match_ID FROM matches"
matches_df = pd.read_sql(query, engine)

# ----------------------------
# RapidAPI headers
# ----------------------------

headers = {
    "x-rapidapi-key": "ece02345femshf1e2084c3c64098p13907bjsn81b907c12629",
    "x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

toss_records = []

# ----------------------------
# Loop through matches
# ----------------------------

for match_id in matches_df["Match_ID"]:

    url = f"https://cricbuzz-cricket.p.rapidapi.com/mcenter/v1/{match_id}"

    response = requests.get(url, headers=headers)

    data = response.json()

    toss_status = data.get("tossstatus")

    if toss_status:

        toss_winner = toss_status.split(" opt")[0]

        decision = "bowl" if "bowl" in toss_status else "bat"

        toss_records.append({
            "Match_ID": match_id,
            "toss_winner_team": toss_winner,
            "decision": decision
        })

        print("Collected:", match_id, toss_winner, decision)

# ----------------------------
# Insert into toss_details table
# ----------------------------

toss_df = pd.DataFrame(toss_records)

toss_df.to_sql("toss_details", engine, if_exists="append", index=False)

print("All toss data inserted successfully")
```
