```python
import requests

# Define a list of team categories to fetch data for
team_categories = ["international", "league", "domestic", "women"]

url_template = "https://cricbuzz-cricket.p.rapidapi.com/teams/v1/{team_type}"

headers = {
	"x-rapidapi-key": "66a5941e36msh6019957b8d66a9bp18647djsnaefa60b20e43",
	"x-rapidapi-host": "cricbuzz-cricket.p.rapidapi.com"
}

# Dictionary to store responses for each team category
vasu_responses = {}

for team_type in team_categories:
    url = url_template.format(team_type=team_type)
    response = requests.get(url, headers=headers)
    # Check if the request was successful before parsing JSON
    if response.status_code == 200:
        vasu_responses[team_type] = response.json()
    else:
        print(f"Error fetching data for {team_type}: {response.status_code} - {response.text}")

# You can now access the data for each category, for example:
# international_teams_data = vasu_responses.get("international")
# print(international_teams_data)

# To display all collected data:
print(vasu_responses)
```

```python
import pymysql
import json

mysql_secrets = {
    "host": "localhost",
    "user": "root",
    "password": "Vasu@76652",
    "database": "cricketdb",
    "port": 3306
}

def _find_teams(resp):
    if isinstance(resp, list):
        return resp
    if not isinstance(resp, dict):
        return None
    # common keys that may contain list of teams
    for key in ('teams','list','data','items','results'):
        val = resp.get(key)
        if isinstance(val, list):
            return val
    # fallback: first list value
    for k,v in resp.items():
        if isinstance(v, list):
            return v
    return None

def insert_team_data(data, secrets):
    try:
        conn = pymysql.connect(
            host=secrets.get('host', 'localhost'),
            user=secrets.get('user'),
            password=secrets.get('password'),
            database=secrets.get('database'),
            port=secrets.get('port', 3306),
            cursorclass=pymysql.cursors.DictCursor,
        )
        cursor = conn.cursor()

        # Connection-level diagnostics
        try:
            cursor.execute('SELECT DATABASE() as db')
            print('Connected DATABASE():', cursor.fetchone())
            cursor.execute('SELECT USER() as user')
            print('Connected USER():', cursor.fetchone())
            cursor.execute('SELECT VERSION() as version')
            print('MySQL VERSION():', cursor.fetchone())
            cursor.execute('SHOW DATABASES')
            dbs = [r for r in cursor.fetchall()]
            print('Databases available (sample 10):', dbs[:10])
            cursor.execute('SHOW TABLES')
            print('Tables in current DB:', [r for r in cursor.fetchall()][:50])
        except Exception as err:
            print('Connection diagnostic queries failed:', err)

        for team_type, resp in data.items():
            teams = _find_teams(resp)
            if not teams:
                print(f"No teams list found for {team_type}; skipping")
                continue
            print(f"Found {len(teams)} teams for {team_type}")
            for team in teams:
                if not isinstance(team, dict):
                    continue
                team_id = team.get('id') or team.get('team_id') or team.get('teamId')
                team_name = team.get('name') or team.get('team_name') or team.get('teamName')
                country = team.get('country') or team.get('country_name') or team.get('countryName')
                # exclude common id/name keys from meta
                meta = json.dumps({k: v for k, v in team.items() if k not in ['id', 'name', 'country', 'teamId', 'team_id', 'teamName', 'team_name', 'countryName', 'country_name']})

                if team_id and team_name:
                    try:
                        insert_query = ('INSERT INTO teams (Team_ID, Team_Name, Country, Meta) VALUES (%s, %s, %s, %s) '
                                        'ON DUPLICATE KEY UPDATE Team_Name=VALUES(Team_Name), Country=VALUES(Country), Meta=VALUES(Meta)')
                        cursor.execute(insert_query, (str(team_id), team_name, country, meta))
                        print(f"Inserted/Updated: {team_name} ({team_id}) | rowcount={cursor.rowcount}")
                    except Exception as err:
                        print(f"Error inserting team {team_name}: {err}")
                else:
                    print(f"Skipping team (missing id/name): id={team.get('teamId') or team.get('team_id') or team.get('id')}, name={team.get('teamName') or team.get('team_name') or team.get('name')}")

        conn.commit()
        print("All data insertion/update complete.")

        # Post-commit verification diagnostics
        try:
            # Fully-qualified count (explicit DB.table)
            fq = f"{secrets.get('database')}.teams"
            cursor.execute(f"SELECT COUNT(*) as cnt FROM `{secrets.get('database')}`.`teams`")
            cnt_row = cursor.fetchone()
            cnt = cnt_row.get('cnt') if cnt_row else None
            print(f"teams table count (post-commit, fully-qualified {fq}): {cnt}")

            # If 0, attempt to SHOW TABLE STATUS for clues
            if not cnt:
                try:
                    cursor.execute('SHOW TABLE STATUS LIKE "teams"')
                    print('SHOW TABLE STATUS LIKE "teams":', cursor.fetchall())
                except Exception as err:
                    print('SHOW TABLE STATUS failed:', err)

            # fetch a few sample rows
            cursor.execute("SELECT Team_ID, Team_Name, Country, Meta FROM teams LIMIT 5")
            samples = cursor.fetchall()
            print("Sample rows:")
            for r in samples:
                print(r)

            # Also try reading via information_schema to see if table exists in that schema
            cursor.execute("SELECT TABLE_SCHEMA, TABLE_NAME FROM information_schema.tables WHERE TABLE_NAME='teams'")
            print('information_schema.tables entries for teams:', cursor.fetchall())

        except Exception as err:
            print(f"Verification query failed: {err}")

    except Exception as err:
        print(f"Error connecting to MySQL: {err}")
    finally:
        try:
            if cursor:
                cursor.close()
            if conn:
                conn.close()
            print("MySQL connection closed.")
        except Exception:
            pass

# Call the function to insert data
insert_team_data(vasu_responses, mysql_secrets)
```
